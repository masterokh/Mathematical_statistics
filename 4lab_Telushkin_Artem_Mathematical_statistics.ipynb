###1. Разумно ли считать, что емкость аккумулятора распределена равномерно?
Формализация задачи:

Проверка гипотезы о равномерности распределения емкости аккумулятора в выборке. Для этого можно воспользоваться критерием Колмогорова-Смирнова и тестом Андерсона-Дарлинга.
Критерий Колмогорова-Смирнова проверяет гипотезу о том, что выборка была взята из некоторого заданного распределения. В данном случае мы можем проверить, что выборка была взята из равномерного распределения. Тест Андерсона-Дарлинга более мощный и может обнаружить не только отклонение от нормальности, но и указать на другие типы отклонений от предполагаемого распределения.

```python
import pandas as pd
import numpy as np
import io
from scipy.stats import uniform, kstest, anderson

# Загрузка данных
from google.colab import files
uploaded = files.upload()
file_name = "mobile_phones.csv"

io.StringIO(uploaded["mobile_phones.csv"].decode("utf-8"))
data = pd.read_csv(io.StringIO(uploaded["mobile_phones.csv"].decode("utf-8")))
print('\n')
data = pd.read_csv('mobile_phones.csv')

# Выбор столбца с емкостью аккумулятора
battery_capacity = data['battery_power']


battery_capacity_all = (battery_capacity - battery_capacity.min()) / (battery_capacity.max() - battery_capacity.min())

# Проверка на равномерность распределения с помощью критерия Колмогорова-Смирнова
D, p_value = kstest(battery_capacity_all, uniform.cdf)
alpha = 0.05
if p_value > alpha:
    print("Распределение емкости аккумулятора можно считать равномерным")
else:
    print("Распределение емкости аккумулятора не является равномерным")


# Проверка на нормальность распределения с помощью теста Андерсона-Дарлинга
result = anderson(battery_capacity)
if result.statistic > result.critical_values[2]:
    print("Распределение емкости аккумулятора можно считать нормальным")
else:
    print("Распределение емкости аккумулятора не является нормальным")
```

###2. Верно ли, что телефонов с поддержкой 3G больше моделей с Wi-Fi? А разнится ли количество телефонов с touch screen от моделей с двумя сим-картами? На каждый вопрос по тесту.

Формализация задачи:

Проверка гипотезы о различии количества телефонов с поддержкой 3G и моделей с Wi-Fi в выборке, а также проверка гипотезы о различии количества телефонов с touch screen и моделей с двумя сим-картами в выборке. Для этого можно воспользоваться критерием хи-квадрат.

Критерий хи-квадрат позволяет проверить гипотезу о том, что распределение двух категориальных переменных связаны между собой. В данном случае мы можем проверить, что есть ли связь между наличием 3G и Wi-Fi у телефонов, а также между наличием touch screen и двумя сим-картами.

```python
import pandas as pd
import io
from scipy.stats import ttest_ind

# Загрузка данных
from google.colab import files
uploaded = files.upload()
file_name = "mobile_phones.csv"
data = pd.read_csv(io.StringIO(uploaded["mobile_phones.csv"].decode("utf-8")))

alpha = 0.05

# Гипотеза: количество телефонов с 3G равно количеству телефонов с Wi-Fi
count_3g = data[data['three_g'] == 1]['three_g'].count()
count_wifi = data[data['wifi'] == 1]['wifi'].count()
t_stat, p_value = ttest_ind(count_3g, count_wifi, alternative='two-sided')

# Проверяем гипотезу
if p_value < alpha:
    print("Отвергаем нулевую гипотезу. Количество телефонов с 3G не равно количеству телефонов с Wi-Fi.")
else:
    print("Не удалось отвергнуть нулевую гипотезу. Количество телефонов с 3G равно количеству телефонов с Wi-Fi.")

# Гипотеза: количество телефонов с touch screen равно количеству телефонов с двумя сим-картами
count_touch_screen = data[data['touch_screen'] == 1]['touch_screen'].count()
count_dual_sim = data[data['dual_sim'] == 1]['dual_sim'].count()
t_stat, p_value = ttest_ind(count_touch_screen, count_dual_sim, alternative='two-sided')

# Проверяем гипотезу
if p_value < alpha:
    print("Отвергаем нулевую гипотезу. Количество телефонов с touch screen не равно количеству телефонов с двумя сим-картами.")
else:
    print("Не удалось отвергнуть нулевую гипотезу. Количество телефонов с touch screen равно количеству телефонов с двумя сим-картами.")
```

###3. Есть подозрение, что цена зависит от объема оперативной памяти. Проверите данное утверждение.

Для проверки гипотезы о линейной зависимости между ценой и объемом оперативной памяти можно воспользоваться коэффициентом корреляции Пирсона и коэффициентом Спирмена.

Коэффициент корреляции Пирсона измеряет степень линейной зависимости между двумя непрерывными переменными. Значения коэффициента корреляции Пирсона лежат в диапазоне от -1 до 1. Значение 0 означает отсутствие корреляции, положительное значение - прямую корреляцию, отрицательное значение - обратную корреляцию. Чем ближе значение коэффициента к 1 или -1, тем более сильная корреляция между переменными.

Коэффициент Спирмена также измеряет степень связи между двумя переменными, но не обязательно линейную. Он работает с рангами значений переменных, а не с их абсолютными значениями. Значения коэффициента Спирмена также лежат в диапазоне от -1 до 1, и их интерпретация аналогична коэффициенту корреляции Пирсона.

```python
import pandas as pd
import io
from scipy.stats import pearsonr, spearmanr

# Загрузка данных
from google.colab import files
uploaded = files.upload()
file_name = "mobile_phones.csv"

io.StringIO(uploaded["mobile_phones.csv"].decode("utf-8"))
data = pd.read_csv(io.StringIO(uploaded["mobile_phones.csv"].decode("utf-8")))
print('\n')
data = pd.read_csv('mobile_phones.csv')

# Проверяем корреляцию Пирсона
corr_coef, p_value = pearsonr(data['price_range'], data['ram'])

if p_value < 0.05:
    print('Отвергаем нулевую гипотезу: между ценой и объемом оперативной памяти есть связь')
else:
    print('Принимаем нулевую гипотезу: между ценой и объемом оперативной памяти нет связи')

# Проверяем корреляцию Спирмена
corr, p_value = spearmanr(data['price_range'], data['ram'])

if p_value < 0.05:
    print("Отвергаем нулевую гипотезу. Существует связь между ценой и объемом оперативной памяти.")
else:
    print("Не отвергаем нулевую гипотезу. Нет связи между ценой и объемом оперативной памяти.")

```

Если значение p-value меньше заданного уровня значимости (меньше 0.05 => верно), то нулевая гипотеза об отсутствии связи между переменными отвергается.

В данном случае, можем судить о наличии сильной положительной линейной связи между ценой и объемом оперативной памяти.
