2. ùúè = ùúá1 ‚àí ùúá2; ùúé1^2 = ùúé2^2 –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–∞; ùúá1 = 2, ùúá2 = 1, ùúé1^2 = ùúé2^2 = 1; –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å —Ñ—É–Ω–∫—Ü–∏–µ–π

((·∫ä ‚àí √ù ‚àí ùúè)/‚àöÔ∏Ä(ùëõVar(ùëã) + ùëöVar(ùëå ))
–≥–¥–µ Var(.) ‚Äì –≤—ã–±–æ—Ä–æ—á–Ω–∞—è —Å–º–µ—â–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è. –°–º–æ—Ç—Ä–∏—Ç–µ –≤ —Å—Ç–æ—Ä–æ–Ω—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞.

```python
import numpy as np
from scipy.stats import f

n = 25
m = 25
alpha = 0.05
num_trials = 1000

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫–∏ X –∏ Y
X = np.random.normal(loc=0, scale=1, size=n)
Y = np.random.normal(loc=0.5, scale=1.5, size=m)

# –í—ã—á–∏—Å–ª—è–µ–º –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ —Å–º–µ—â–µ–Ω–Ω—ã–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
var_X = np.var(X, ddof=1)
var_Y = np.var(Y, ddof=1)

# –í—ã—á–∏—Å–ª—è–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –¥–ª—è –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ mu_1 - mu_2
t = (np.mean(X) - np.mean(Y)) / np.sqrt(var_X/n + var_Y/m)

# –í—ã—á–∏—Å–ª—è–µ–º –∫–≤–∞–Ω—Ç–∏–ª—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –°—Ç—å—é–¥–µ–Ω—Ç–∞
q = f.ppf(1 - alpha/2, dfn=n-1, dfd=m-1)

# –í—ã—á–∏—Å–ª—è–µ–º –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª
ci_low = t - q*np.sqrt((n-1)*var_X/n + (m-1)*var_Y/m)/(np.sqrt(n+m-2)*np.sqrt(1/n + 1/m))
ci_high = t + q*np.sqrt((n-1)*var_X/n + (m-1)*var_Y/m)/(np.sqrt(n+m-2)*np.sqrt(1/n + 1/m))

# –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –ø–æ–∫—Ä—ã–≤–∞–µ—Ç –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞
num_covered = 0
for i in range(num_trials):
 X = np.random.normal(loc=0, scale=1, size=n)
 Y = np.random.normal(loc=0.5, scale=1.5, size=m)
 var_X = np.var(X, ddof=1)
 var_Y = np.var(Y, ddof=1)
 t = (np.mean(X) - np.mean(Y)) / np.sqrt(var_X/n + var_Y/m)
 q = f.ppf(1 - alpha/2, dfn=n-1, dfd=m-1)
 ci_low = t - q*np.sqrt((n-1)*var_X/n + (m-1)*var_Y/m)/(np.sqrt(n+m-2)*np.sqrt(1/n + 1/m))
 ci_high = t + q*np.sqrt((n-1)*var_X/n + (m-1)*var_Y/m)/(np.sqrt(n+m-2)*np.sqrt(1/n + 1/m))
 if ci_low <= 0 <= ci_high:
  num_covered += 1

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
coverage_rate = num_covered / num_trials
print(f'–î–ª—è –≤—ã–±–æ—Ä–∫–∏ –æ–±—ä–µ–º–∞ {n} –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –ø–æ–∫—Ä—ã–≤–∞–µ—Ç —Ä–µ–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –≤ {coverage_rate:.3f} —Å–ª—É—á–∞—è—Ö')

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫–∏
n = 10000
m = 10000

coverage = 0
for i in range(num_trials):
    X = np.random.normal(loc=0, scale=1, size=n)
    Y = np.random.normal(loc=0, scale=1, size=m)

    # –°—á–∏—Ç–∞–µ–º –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
    var_X = np.var(X, ddof=1)
    var_Y = np.var(Y, ddof=1)

    # –°—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É –∏ –∫–≤–∞–Ω—Ç–∏–ª–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –§–∏—à–µ—Ä–∞
    statistic = (np.mean(X) - np.mean(Y)) / np.sqrt(var_X/n + var_Y/m)
    q1 = f.ppf(alpha/2, n-1, m-1)
    q2 = f.ppf(1 -alpha/2, n-1, m-1)

    # –°—á–∏—Ç–∞–µ–º –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª
    ci_low = statistic/q2
    ci_high = statistic/q1

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –ø–æ–ø–∞–¥–∞–µ—Ç –ª–∏ –∏—Å—Ç–∏–Ω–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª
    if ci_low <= 1 <= ci_high:
     num_covered += 1

print(f'–î–ª—è –≤—ã–±–æ—Ä–∫–∏ –æ–±—ä–µ–º–∞ {n} –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –ø–æ–∫—Ä—ã–≤–∞–µ—Ç —Ä–µ–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ –≤ {num_covered:.3f} —Å–ª—É—á–∞—è—Ö')
```

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –≥—Ä–∞–Ω–∏—Ü—ã –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è –∫–∞–∫:

[sample_mean - z * sample_std / sqrt(n), sample_mean + z * sample_std / sqrt(n)]

–≥–¥–µ sqrt - —Ñ—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –∫–æ—Ä–Ω—è.

```python
import numpy as np
from scipy import stats

p = 0.7 # –ø–∞—Ä–∞–º–µ—Ç—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
n = 100 # —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏
alpha = 0.05 # —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏

sample = np.random.binomial(1, p, n) # –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É
sample_mean = np.mean(sample) # –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ
sample_std = np.std(sample, ddof=1) # —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ

z = np.abs(stats.norm.ppf(alpha / 2)) # –∫–≤–∞–Ω—Ç–∏–ª—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
interval = (sample_mean - z * sample_std / np.sqrt(n), sample_mean + z * sample_std / np.sqrt(n))

print(f"–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª: [{interval[0]:.4f}, {interval[1]:.4f}]")
```
