###–î–ª—è –Ω–∞—á–∞–ª–∞ –Ω–∞–π–¥—ë–º –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –æ–±—ä–µ–º –≤—ã–±–æ—Ä–∫–∏ n:

ùëÉ(|ùëãÃÑ ‚àí ùúáùúÉ| ‚â• ùúÄ) ‚â§ ùúé¬≤/ (ùúÄ¬≤ùëõ)

–≥–¥–µ ùëãÃÑ - –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ, ùúáùúÉ - –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ, ùúé¬≤ - –¥–∏—Å–ø–µ—Ä—Å–∏—è.

–¢–∞–∫ –∫–∞–∫ —è –Ω–µ –∑–Ω–∞—é –¥–∏—Å–ø–µ—Ä—Å–∏—é, —Ç–æ –±—É–¥—É –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Å –ø–æ–º–æ—â—å—é –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏:

ùëÜ¬≤ = 1/(ùëõ ‚àí 1) ‚àë(ùëãùëñ ‚àí ùëãÃÑ)¬≤

–ü–µ—Ä–µ–ø–∏—à—É –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –ß–µ–±—ã—à–µ–≤–∞ —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:

ùëÉ(|ùëãÃÑ ‚àí ùúáùúÉ| ‚â• ùúÄ) ‚â§ ùëÜ¬≤/(ùúÄ¬≤ùëõ)

–ù—É–∂–Ω–æ —á—Ç–æ–±—ã —ç—Ç–æ –Ω–µ—Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –±—ã–ª–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–æ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –Ω–µ –º–µ–Ω—å—à–µ–π 1‚àíùõø.

ùëÜ¬≤/(ùúÄ¬≤ùëõ) ‚â§ 1 ‚àí ùõø

ùëõ ‚â• ùëÜ¬≤/(ùúÄ¬≤(1 ‚àí ùõø))

–ü–æ–¥—Å—Ç–∞–≤–ª—é –∫–æ–Ω–∫—Ä–µ—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–∏:

ùëù = 4/5

ùúáùúÉ = 1/ùëù = 5/4

ùúé¬≤ = (1 ‚àí ùëù)/ùëù¬≤ = 1/16

ùúÄ = 0.01

ùõø = 0.05

–¢–æ–≥–¥–∞, –ø–æ–¥—Å—Ç–∞–≤–ª—è—è –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Ñ–æ—Ä–º—É–ª—É, –ø–æ–ª—É—á–∞—é:

ùëõ ‚â• 1/16/(0.01¬≤(1‚àí0.05)) ‚âà 3136

–Ø –¥–æ–ª–∂–µ–Ω –≤–∑—è—Ç—å –≤—ã–±–æ—Ä–∫—É —Ä–∞–∑–º–µ—Ä–æ–º –Ω–µ –º–µ–Ω–µ–µ 3136, —á—Ç–æ–±—ã –≥–∞—Ä–∞–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å, —á—Ç–æ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è –Ω–µ –±–æ–ª–µ–µ —á–µ–º –Ω–∞ 0.01 —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –Ω–µ –º–µ–Ω—å—à–µ–π 0.95.

–¢–µ–ø–µ—Ä—å —Å–≥–µ–Ω–µ—Ä–∏—Ä—É—é 500 –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–∑–º–µ—Ä–æ–º 3136 –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è Geom(ùëù) —Å ùëù = 4/5 –∏ –ø–æ—Å—á–∏—Ç–∞—é, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è –±–æ–ª–µ–µ —á–µ–º –Ω–∞ 0.01.  

###–î–∞–ª–µ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω –∫–æ–¥ –Ω–∞ Python:

```python
import numpy as np
from scipy.stats import geom

# –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –∑–∞–¥–∞—á–∏
p = 4/5
mu = 1/p
epsilon = 0.01
delta = 0.05
n = 3136
num_samples = 500

# –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫–∏ –∏ –ø–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑, –∫–æ–≥–¥–∞ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏—è –±–æ–ª–µ–µ —á–µ–º –Ω–∞ epsilon
count = 0
for i in range(num_samples):
    sample = geom.rvs(p, size=n)
    sample_mean = np.mean(sample)
    if abs(sample_mean - mu) > epsilon:
        count += 1

print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è –æ—Ç –º–∞—Ç. –æ–∂–∏–¥–∞–Ω–∏—è –±–æ–ª–µ–µ —á–µ–º –Ω–∞ {epsilon}: {count}")
```
